Data collection using Webscraping:

```python
import requests
from bs4 import BeautifulSoup
import csv
import json
import time

def get_additional_details(restaurant_url):
    try:
        response = requests.get(restaurant_url)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')

        other_info_section = soup.find('dd', class_='other-information')

        details = {
            'cuisine': 'Cuisine type not found',
            'price_range': 'Price range not found'
        }

        if other_info_section:
            info_text = other_info_section.get_text(strip=True)
            parts = info_text.split('Price Range')

            if len(parts) > 1:
                details['cuisine'] = parts[0].replace('Cuisines:', '').strip()
                details['price_range'] = parts[1].strip()
            else:
                details['cuisine'] = info_text.replace('Cuisines:', '').strip()

        return details
    except requests.exceptions.RequestException as e:
        print(f"Error fetching URL {restaurant_url}: {e}")
        return {'cuisine': 'Error retrieving cuisine', 'price_range': 'Error retrieving price range'}

# Base URL of the page to scrape
base_url = 'https://www.yellowpages.com/dallas-tx/restaurants'

# Initialize page number
page_number = 1
max_pages = 15 # Maximum number of pages to scrape

# Lists to store data
data = []

# Loop to go through multiple pages
while page_number <= max_pages:
    # Create URL for each page
    url = f'{base_url}?page={page_number}'

    # Send a request to the URL
    response = requests.get(url)

    # Break the loop if the page doesn't exist or has no results
    if response.status_code != 200 or "Sorry" in response.text:
        break

    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all restaurant listings
    restaurants = soup.find_all('div', class_='info')

    # Break the loop if no restaurants are found (end of results)
    if not restaurants:
        break

    # Extract and store restaurant details
    for restaurant in restaurants:
        # Restaurant name
        name = restaurant.find('a', class_='business-name')
        name = name.text.strip() if name else 'No name available'

        # Phone number
        phone = restaurant.find('div', class_='phones phone primary')
        phone = phone.text.strip() if phone else 'No phone number available'

        # Address
        address = restaurant.find('div', class_='street-address')
        address = address.text.strip() if address else 'No address available'

        # City, state, and zip
        city_state_zip = restaurant.find('div', class_='locality')
        city_state_zip = city_state_zip.text.strip() if city_state_zip else 'No city/state/zip available'

        # Years in business
        years = restaurant.find('div', class_='years-in-business')
        years = years.text.strip() if years else 'No data'

        # Secondary info
        secondary = restaurant.find('div', class_='snippet')
        secondary_info = secondary.text.strip() if secondary else 'No secondary info'

        # Categories
        category = restaurant.find('div', class_='categories')
        category = category.text.strip() if category else 'No category'

        # Ratings and TripAdvisor details
        rating_div = restaurant.find('div', class_='ratings')
        tripadvisor_rating = 'No TripAdvisor rating available'
        tripadvisor_rating_count = 'No TripAdvisor rating count available'
        displayed_rating = 'No displayed rating'
        rating_count = 'No rating count available'

        if rating_div:
            # Extract TripAdvisor rating from data-tripadvisor attribute
            rating_data = rating_div.get('data-tripadvisor')
            if rating_data:
                try:
                    rating_data = json.loads(rating_data.replace('&quot;', '"'))
                    tripadvisor_rating = rating_data.get('rating', 'No TripAdvisor rating available')
                    tripadvisor_rating_count = rating_data.get('count', 'No TripAdvisor rating count available')
                except json.JSONDecodeError:
                    pass

            # Extract displayed rating from classes
            rating_span = rating_div.find('div', class_='result-rating')
            if rating_span:
                classes = rating_span.get('class', [])
                for cls in classes:
                    if 'five' in cls:
                        displayed_rating = 5.0
                    elif 'four half' in cls:
                        displayed_rating = 4.5
                    elif 'four' in cls:
                        displayed_rating = 4.0
                    elif 'three half' in cls:
                        displayed_rating = 3.5
                    elif 'three' in cls:
                        displayed_rating = 3.0
                    elif 'two half' in cls:
                        displayed_rating = 2.5
                    elif 'two' in cls:
                        displayed_rating = 2.0
                    elif 'one half' in cls:
                        displayed_rating = 1.5
                    elif 'one' in cls:
                        displayed_rating = 1.0
                    # Note: Add handling for other potential classes if needed

            # Extract rating count if available
            count_span = rating_div.find('span', class_='count')
            rating_count = count_span.text.strip() if count_span else 'No rating count available'

        # Extract restaurant page URL
        link = restaurant.find('a', class_='business-name')
        restaurant_url = 'https://www.yellowpages.com' + link['href'] if link and link['href'] else 'No URL available'

        # Get additional details from the restaurant's page
        additional_details = get_additional_details(restaurant_url)

        # Add the data to the list
        data.append({
            'Name': name,
            'Phone': phone,
            'Address': address,
            'City/State/Zip': city_state_zip,
            'Years in Business': years,
            'Secondary Info': secondary_info,
            'Category': category,
            'Displayed Rating': displayed_rating,
            'Rating Count': rating_count,
            'TripAdvisor Rating': tripadvisor_rating,
            'TripAdvisor Rating Count': tripadvisor_rating_count,
            'Cuisine': additional_details['cuisine'],
            'Price Range': additional_details['price_range'],
            'URL': restaurant_url
        })

    # Print the page number (for debugging purposes)
    print(f"Scraping page {page_number}")

    # Increment the page number
    page_number += 1

    # Optional: Add a delay to avoid being blocked
    time.sleep(1)

# Write the data to a CSV file
with open('full_restaurant_data.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = [
        'Name', 'Phone', 'Address', 'City/State/Zip', 'Years in Business',
        'Secondary Info', 'Category', 'Displayed Rating', 'Rating Count',
        'TripAdvisor Rating', 'TripAdvisor Rating Count', 'Cuisine',
        'Price Range', 'URL'
    ]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    writer.writerows(data)

print("Full restaurant data has been written to full_restaurant_data.csv")
```

```python
import re

# Function to extract numeric value from text
def extract_numeric(text):
    if pd.isna(text):
        return pd.NA
    return pd.to_numeric(re.sub(r'[^\d]', '', text), errors='coerce')

# Clean 'Years in Business'
data['Years in Business'] = data['Years in Business'].apply(lambda x: extract_numeric(x))

# Clean 'Rating Count'
data['Rating Count'] = data['Rating Count'].apply(lambda x: extract_numeric(x))

# Clean 'TripAdvisor Rating'
data['TripAdvisor Rating'] = data['TripAdvisor Rating'].apply(lambda x: pd.to_numeric(x, errors='coerce'))

# Clean 'TripAdvisor Rating Count'
data['TripAdvisor Rating Count'] = data['TripAdvisor Rating Count'].apply(lambda x: extract_numeric(x))

# Handle missing values: Fill or remove as necessary
data['Rating Count'] = data['Rating Count'].fillna(0)
data['TripAdvisor Rating Count'] = data['TripAdvisor Rating Count'].fillna(0)

# Display the cleaned DataFrame
print(data.head())
```

Outliers:

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

# Apply to relevant columns
data = remove_outliers(data, 'Years in Business')
data = remove_outliers(data, 'Rating Count')
data = remove_outliers(data, 'TripAdvisor Rating')
data = remove_outliers(data, 'TripAdvisor Rating Count')
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
import pandas as pd


# Define features and target variable
X = data[['Years in Business', 'Displayed Rating', 'TripAdvisor Rating', 'Rating Count', 'Cuisine']]
y = data['Price Range']

# Define categorical and numerical columns
categorical_features = ['Cuisine']
numerical_features = ['Years in Business', 'Displayed Rating', 'TripAdvisor Rating', 'Rating Count']

# Create transformers for preprocessing
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ]
)

# Apply transformations
X_preprocessed = preprocessor.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y, test_size=0.2, random_state=42)
```

```python
# Check unique class labels in the target variable
unique_classes = y.unique()
print("Unique classes in target variable:", unique_classes)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# Initialize and train the model
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)

# Predict and evaluate
y_pred = logreg.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Determine unique classes and adjust class names accordingly
unique_classes = sorted(set(y_test) | set(y_pred))
class_names = [f'Class {i}' for i in unique_classes]  # Create class names dynamically

# Generate classification report
report = classification_report(y_test, y_pred, target_names=class_names)

# Generate confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Print results
print("Logistic Regression:")
print(f"Accuracy: {accuracy:.4f}\n")
print("Classification Report:\n", report)
print("Confusion Matrix:\n", conf_matrix)
```
